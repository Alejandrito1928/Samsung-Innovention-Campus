```python
# Usamos esto para que cada pieza tenga un número diferente y poder verlas
tromino_id = 0

def imprimir_tablero(tablero):
    """Esta función solo sirve para dibujar el resultado final de forma bonita."""
    for fila in tablero:
        print(" ".join(f"{str(celda):>2}" for celda in fila))

# --- LA FUNCIÓN "MANAGER" ---
# A esta función le das una sección del tablero y la casilla que no debe tocar.
def resolver_tromino(tablero, x_sup_izq, y_sup_izq, tam, x_hueco, y_hueco):
    global tromino_id

    # --- El Caso Sencillo (La Tarea Fácil) ---
    # Si la tarea es un tablero de 2x2, es muy fácil.
    # Se pone una pieza y se termina el trabajo.
    if tam == 2:
        tromino_id += 1
        for i in range(tam):
            for j in range(tam):
                if tablero[x_sup_izq + i][y_sup_izq + j] == 0: # Si la casilla está libre
                    tablero[x_sup_izq + i][y_sup_izq + j] = tromino_id
        return # ¡Trabajo hecho!

    # --- El Caso Complicado (La Estrategia del Manager) ---
    
    # 1. Dividir la tarea en 4 partes
    medio = tam // 2
    tromino_id += 1
    
    # 2. Poner el trominó del centro para conectar a los equipos.
    # El manager mira dónde está el hueco para saber cómo poner la pieza central.
    
    # Si el hueco está en el CUADRANTE 1 (arriba-izquierda)
    if x_hueco < x_sup_izq + medio and y_hueco < y_sup_izq + medio:
        # Pone la pieza central ocupando una casilla de los otros 3 cuadrantes.
        tablero[x_sup_izq + medio - 1][y_sup_izq + medio] = tromino_id
        tablero[x_sup_izq + medio][y_sup_izq + medio - 1] = tromino_id
        tablero[x_sup_izq + medio][y_sup_izq + medio] = tromino_id
        
        # 3. Delegar cada parte a un "sub-manager" (llamada recursiva)
        # El manager le dice al primer sub-manager que resuelva el cuadrante 1,
        # indicando dónde está el hueco original.
        resolver_tromino(tablero, x_sup_izq, y_sup_izq, medio, x_hueco, y_hueco)
        # A los otros 3 sub-managers les dice que su "hueco" es la pieza que él acaba de poner.
        resolver_tromino(tablero, x_sup_izq, y_sup_izq + medio, medio, x_sup_izq + medio - 1, y_sup_izq + medio)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq, medio, x_sup_izq + medio, y_sup_izq + medio - 1)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq + medio, medio, x_sup_izq + medio, y_sup_izq + medio)

    # ... y hace lo mismo para los otros 3 casos posibles (si el hueco está en otro cuadrante) ...
    # (El resto del código maneja esas otras 3 posibilidades, pero la lógica es idéntica)

    elif x_hueco < x_sup_izq + medio and y_hueco >= y_sup_izq + medio: # Hueco en CUADRANTE 2 (arriba-derecha)
        tablero[x_sup_izq + medio - 1][y_sup_izq + medio - 1] = tromino_id
        tablero[x_sup_izq + medio][y_sup_izq + medio - 1] = tromino_id
        tablero[x_sup_izq + medio][y_sup_izq + medio] = tromino_id
        
        resolver_tromino(tablero, x_sup_izq, y_sup_izq, medio, x_sup_izq + medio - 1, y_sup_izq + medio - 1)
        resolver_tromino(tablero, x_sup_izq, y_sup_izq + medio, medio, x_hueco, y_hueco)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq, medio, x_sup_izq + medio, y_sup_izq + medio - 1)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq + medio, medio, x_sup_izq + medio, y_sup_izq + medio)
    
    elif x_hueco >= x_sup_izq + medio and y_hueco < y_sup_izq + medio: # Hueco en CUADRANTE 3 (abajo-izquierda)
        tablero[x_sup_izq + medio - 1][y_sup_izq + medio - 1] = tromino_id
        tablero[x_sup_izq + medio - 1][y_sup_izq + medio] = tromino_id
        tablero[x_sup_izq + medio][y_sup_izq + medio] = tromino_id
        
        resolver_tromino(tablero, x_sup_izq, y_sup_izq, medio, x_sup_izq + medio - 1, y_sup_izq + medio - 1)
        resolver_tromino(tablero, x_sup_izq, y_sup_izq + medio, medio, x_sup_izq + medio - 1, y_sup_izq + medio)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq, medio, x_hueco, y_hueco)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq + medio, medio, x_sup_izq + medio, y_sup_izq + medio)
        
    else: # Hueco en CUADRANTE 4 (abajo-derecha)
        tablero[x_sup_izq + medio - 1][y_sup_izq + medio - 1] = tromino_id
        tablero[x_sup_izq + medio - 1][y_sup_izq + medio] = tromino_id
        tablero[x_sup_izq + medio][y_sup_izq + medio - 1] = tromino_id

        resolver_tromino(tablero, x_sup_izq, y_sup_izq, medio, x_sup_izq + medio - 1, y_sup_izq + medio - 1)
        resolver_tromino(tablero, x_sup_izq, y_sup_izq + medio, medio, x_sup_izq + medio - 1, y_sup_izq + medio)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq, medio, x_sup_izq + medio, y_sup_izq + medio - 1)
        resolver_tromino(tablero, x_sup_izq + medio, y_sup_izq + medio, medio, x_hueco, y_hueco)


# --- EL "GRAN JEFE" QUE EMPIEZA TODO ---
if __name__ == "__main__":
    # Definimos el tamaño del tablero y la posición del hueco
    N = 8
    HUECO_X, HUECO_Y = 3, 5
    
    # Preparamos el tablero
    tablero = [[0] * N for _ in range(N)]
    tablero[HUECO_X][HUECO_Y] = ' X' # Ponemos el "post-it"
    
    # El gran jefe llama al primer manager para que resuelva todo el tablero
    resolver_tromino(tablero, 0, 0, N, HUECO_X, HUECO_Y)
    
    # Imprimimos el resultado final del trabajo
    imprimir_tablero(tablero)
```

```python
import random

def crear_tablero(x):
    tablero = [[0 for _ in range(x)] for _ in range(x)]
    
    posicion_x = random.randint(0, x - 1)
    posicion_y = random.randint(0, x - 1)
    tablero[posicion_x][posicion_y] = "X"
    
    return tablero

def resolver_tromino(tablero_a_resolver):
    tamano = len(tablero_a_resolver)
    
    for i in range(tamano):
        for j in range(tamano):
            if tablero_a_resolver[i][j] == 0:
                tablero_a_resolver[i][j] = 1
    
    return tablero_a_resolver

def imprimir_tablero(tablero_a_imprimir):
    for fila in tablero_a_imprimir:
        print(fila)

mi_tablero = crear_tablero(4)

print(" Tablero Inicial ")
imprimir_tablero(mi_tablero)

tablero_final = resolver_tromino(mi_tablero)

print("\n Tablero Resuelto ")
imprimir_tablero(tablero_final)
```

```python
def maximun_array(S):
    if len(S) == 0:
        return 0
    elif len(S) == 1:
        return S[0]
    elif len(S) == 2:
        return S[0] + S[1]
    else:
        return "La lógica para listas más grandes aún no está implementada."

S = [2, 3]
M = maximun_array(S)
print(f"Para la lista {S}, el resultado es: {M}")

S_vacia = []
M_vacio = maximun_array(S_vacia)
print(f"Para la lista {S_vacia}, el resultado es: {M_vacio}")

S_uno = [-5]
M_uno = maximun_array(S_uno)
print(f"Para la lista {S_uno}, el resultado es: {M_uno}")
```

```python
def maximum_subarray(nums):
    if not nums:
        return 0

    max_actual = nums[0]
    max_global = nums[0]

    for numero in nums[1:]:
        max_actual = max(numero, max_actual + numero)

        if max_actual > max_global:
            max_global = max_actual
            
    return max_global

S = [-2, 1, -3, 4, -1, 2, 1, -5, 7]
M = maximum_subarray(S)
print(M)
```
