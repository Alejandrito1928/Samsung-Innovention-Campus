```python
import pandas as pd
import numpy as np

# Crear una Serie desde una lista (índice automático)
serie_lista = pd.Series([100, 200, 300])
print("--- Serie desde Lista ---")
print(serie_lista)

# Crear una Serie desde un diccionario (claves = índice)
datos_dict = {'a': 10, 'b': 20, 'c': 30}
serie_dict = pd.Series(datos_dict)
print("\n--- Serie desde Diccionario ---")
print(serie_dict)
```

```python
# Usando la Serie del diccionario anterior
# Por etiqueta de índice
print(f"\nValor en 'b': {serie_dict['b']}")

# Por posición (aunque tenga etiquetas, la posición sigue existiendo)
print(f"Valor en posición 0: {serie_dict[0]}")

# Slicing (corte) por posición
print("\nSlicing [0:2]:")
print(serie_lista[0:2])
```

```python
s1 = pd.Series([10, 20], index=['a', 'b'])
s2 = pd.Series([5, 50], index=['a', 'c']) # Nota: 'b' falta, 'c' es nuevo

# 'a' coincide (10 + 5)
# 'b' no está en s2 -> NaN
# 'c' no está en s1 -> NaN
suma = s1 + s2

print("\n--- Suma de Series (con NaN) ---")
print(suma)
```

```python
datos_stats = pd.Series([1, 5, 10, np.nan, 2]) # Incluimos un NaN

print("\n--- Estadísticas Descriptivas ---")
# .count() ignora los NaN
print(f"Conteo (ignora NaN): {datos_stats.count()}") # [cite: 1076, 1111]
# .mean() también ignora NaN por defecto
print(f"Promedio (ignora NaN): {datos_stats.mean()}") # [cite: 1080]
print(f"Suma (ignora NaN): {datos_stats.sum()}") # [cite: 1078, 1116]
print(f"Máximo: {datos_stats.max()}") # [cite: 1090]
```

```python
import matplotlib.pyplot as plt

# Gráfico de Línea (para tendencias)
# plt.plot() asume el índice como X y los valores como Y
serie_linea = pd.Series([1, 7, 2, 5])
plt.figure() # Crea una nueva figura
plt.plot(serie_linea)
plt.title("Gráfico de Línea Básico")
plt.xlabel("Índice (automático)")
plt.ylabel("Valores")
plt.show() # Muestra el gráfico

# Gráfico de Barras Horizontales (para comparar categorías)
# plt.barh() es mejor para índices con nombres largos
serie_barras = pd.Series([50, 40, 30], index=['Manzanas', 'Naranjas', 'Mangos'])
plt.figure()
plt.barh(serie_barras.index, serie_barras.values) # [cite: 1175]
plt.title("Gráfico de Barras Horizontales")
plt.xlabel("Cantidad")
plt.show()
```
