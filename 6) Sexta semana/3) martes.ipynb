```python
import pandas as pd
s = pd.Series([10, 20, 30, 40, 50],index=['a', 'b', 'c', 'd', 'e'])
s2 = pd.Series([100, 200, 300, 400, 500],index=['a', 'b', 'c', 'd', 'e'])

data = {"one":s,"two":s2}

pd.DataFrame(data)
# This code creates a pandas DataFrame from two Series objects. Each Series
# represents a column in the DataFrame, with the specified indices serving 
# as the row labels. The resulting DataFrame will have two columns 
# named "one" and "two", containing the values from the respective Series.

```

```python
import pandas as pd 

d ={"col1": [1,2,3,4,5],
    "col2": [10,20,30,40,50]}

df = pd.DataFrame(d)
print(df)
# This code creates a pandas DataFrame from a dictionary where the keys
# represent column names and the values are lists of data for each column.

df = df.rename(columns={"col1":"primera_columna","col2":"segunda_columna"})
print(df)

df.rename(index={0:"primera_fila",1:"segunda_fila",2:"tercera_fila",3:"cuarta_fila",4:"quinta_fila"})
# This code renames the columns of the DataFrame to "primera_columna" and "segunda_columna".
# It also renames the index labels to "primera_fila", "segunda_fila", "tercera_fila", "cuarta_fila", and "quinta_fila".
```
