### 2. Creating a DataFrame

```python
import pandas as pd
data = {'col_1': [3, 2, 1, 0], 'col_2': ['a', 'D', 'c', 'd']}
pd.DataFrame(data)
```

```python
import pandas as pd
s = pd.Series([1.0, 2.0, 3.0], index=["a", "b", "c"])
s2 = pd.Series ([1.0, 2.0, 3.0, 4.0], index=["a" , "b", "c", "d"])

data = {"one":s, "two":s2}

pd.DataFrame(data)
```

```python
import pandas as pd
d = {'coll': [1, 2], 'col2': [3, 4]}
df = pd. DataFrame(data=d)
print(df)

df.index = ["new index0", "new index1"]
print(df)

df.columns = ["new_col1", "new_col2"]

print(df)
```

```python
import pandas as pd
d = {'coll': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data=d)
print(df)

df.rename(index= {0:"new index0"})
```

```python
df.rename(columns ={"col2":"new col2"})
```

```python
import pandas as pd
d = {'coll': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data=d)
df.rename(index= {0:"new index0"}, inplace=True)
df.rename(columns ={"col2": "new col2"})
```

```python
import pandas as pd
s= pd.Series([1.0, 2.0, 3.0], index=["a", "b", "c"])
s2= pd.Series([1.0, 2.0, 3.0, 4.0], index=["a", "b", "c", "d"])

dic_data = {"one":s, "two":s2}

type(dic_data)

pd.DataFrame(dic_data)

pd.DataFrame.from_dict(dic_data, orient = 'columns')
```

```python
pd.DataFrame.from_dict(dic_data, orient = 'index')
```

```python
data = {'row_1': [3, 2, 1, 0], 'row_2': ['a', 'b', 'c', 'd']}
pd.DataFrame.from_dict(data, orient='index', columns=['A', 'B', 'C', 'D'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = song[["Rank", "Artist", "Track Name", "Artist_popularity", "Streams", "tempo", "danceability", "energy"]]
df.head()
```

```python
pop_song = df[df["Artist_popularity"]>90]
pop_song.head()
```
